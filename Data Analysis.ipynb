```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_url = "https://drive.google.com/uc?id=15V5a2jEV1-lnAnitcUrJ4rf0dOdu8HnH"
# Read the data into a pandas DataFrame
df = pd.read_csv(file_url)
print(df.head())
```

```python
# Split the data into separate DataFrames 
df_50 = df[df['Edge devices count'] == 50]
df_100 = df[df['Edge devices count'] == 100]
df_200 = df[df['Edge devices count'] == 200]

print(df_50)
```

```python
# Filter the relevant columns for analysis
relevant_cols = ['Orchestration algorithm', 'Total tasks execution delay (s)',
                 'Total tasks waiting time (s)', 'Total tasks executed (Cloud)',
                 'Tasks successfully executed (Cloud)', 'Total tasks executed (Edge)',
                 'Tasks successfully executed (Edge)', 'Cloud energy consumption (Wh)',
                 'Edge energy consumption (Wh)']

df_analysis_50 = df_50[relevant_cols]
df_analysis_100 = df_100[relevant_cols]
df_analysis_200 = df_200[relevant_cols]
```

```python
# Calculate the mean values for relevant metrics for 50 and 100 edge devices
df_mean_50 = df_analysis_50.groupby('Orchestration algorithm').mean().reset_index()
df_mean_100 = df_analysis_100.groupby('Orchestration algorithm').mean().reset_index()
df_mean_200 = df_analysis_200.groupby('Orchestration algorithm').mean().reset_index()   

print(df_mean_50)
```

```python
# Plotting comparison between Edge and Cloud for total tasks executed (50 Edge Devices)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_mean_50, x='Orchestration algorithm', y='Total tasks executed (Edge)', color='blue', label='Edge (50 Edge Devices)')
sns.barplot(data=df_mean_50, x='Orchestration algorithm', y='Total tasks executed (Cloud)', color='orange', label='Cloud (50 Edge Devices)')
plt.title('Total Tasks Executed in Edge vs. Cloud (50 Edge Devices)')
plt.xlabel('Orchestration Algorithm')
plt.ylabel('Total Tasks Executed')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Execution Location', loc='upper left')
plt.tight_layout()
plt.show()

# Print text results for Tasks Executed (50 Edge Devices)
print("Tasks Executed Plot (50 Edge Devices):\n")
for index, row in df_mean_50.iterrows():
    print(f"{row['Orchestration algorithm']}:")
    print(f"- Edge: {row['Total tasks executed (Edge)']} tasks")
    print(f"- Cloud: {row['Total tasks executed (Cloud)']} tasks\n")
```
